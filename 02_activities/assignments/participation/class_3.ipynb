```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(613)
x = np.arange(50)
y1= np.random.randint(0,100,50)
y2= np.random.randint(0,100,50)
```

### Plot both y vars on the same axis
- Add Legend

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label='Person 1', color='orange')
ax.plot(x, y2, label='Person 2', color='grey')
ax.legend(loc = 'lower right')
ax.set_title('Multiple Line Charts')
ax.set_xlabel('x')
ax.set_ylabel('y')
# fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label='Person 1', color='orange')
ax.plot(x, y2, label='Person 2', color='grey')
ax.legend(loc = 'lower right',
          frameon = True,
          fontsize =12,
          ncol =2,
          shadow = True)
ax.set_title('Multiple Line Charts')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

- Adjust the position of the text()

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1', color='orange')
ax.scatter(x, y2, label='Person 2', color='grey')
ax.legend(loc = 'lower right')
ax.text(10,95, "This value is important!",
        ha = 'center',
        color ='red',
        size =20)
# (10,95) will be the center of the text
ax.set_title('Multiple Line Charts')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
```

```python
styles = plt.style.available
n_cols = 5
n_rows = (len(styles) + n_cols - 1) // n_cols 

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20, n_rows * 4))
if len(styles) == 1:
    axes = [axes]
elif len(styles) <= n_cols:
    axes = axes.reshape(-1)  
else:
    axes = axes.flatten() 

for ax, style in zip(axes, styles):
    plt.style.use(style)  
    ax.plot(x, y1)
    ax.plot(x, y2)
    ax.set_title(f'Style: {style}', fontsize=10)
    # ax.grid(True)  


for ax in axes[len(styles):]:
    ax.axis('off')  

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.tight_layout()
```

- Modify each subplot

```python
fig, (ax1, ax2) = plt.subplots(ncols =2,
                              nrows =1,
                               figsize =(7,3) )
plt.style.use('ggplot')
ax1.scatter(x1,y1)
ax1.plot(x1,y1)
ax1.set_title("Person 1")
ax1.legend(loc = 'lower right',
          frameon = True,
          fontsize =12,
          ncol =2,
          shadow = True)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

plt.style.use('fast')
ax2.bar(x2,y2)
```

- Subplots without a grid arrangement

```python
fig, axes = plt.subplot_mosaic([['ax1','ax3'],
                                ['ax2','ax3']
                                ],
                                figsize = (7,4))
axes['ax1'].scatter(x1,y1)
axes['ax2'].bar(x2,y2)
axes['ax3'].plot(x1,y1)
```

- Adjust the space among axes
    - Using layout ='constrained'

```python
fig, axes = plt.subplot_mosaic([['ax1','ax3'],
                                ['ax2','ax3']
                                ],
                                figsize = (7,4),
                                layout='constrained')
axes['ax1'].scatter(x1,y1)
axes['ax2'].bar(x2,y2)
axes['ax3'].plot(x1,y1)

# plt.subplots_adjust(wspace=0.3, hspace=0.3)
axes['ax1'].set_xlabel("A Big Label!", fontsize =18)
axes['ax2'].set_xlabel("Another Label!", fontsize =18)
axes['ax3'].set_xlabel("Label2: fast able to label!", fontsize =18)
```

- Error Handling

```python
y2_sd = np.std(y2)
print(y2_sd)
fig, ax = plt.subplots(figsize =(7,3))
ax.plot(x,y2, color ='red')

ax.errorbar(x, 
            y2,
            yerr= y2_sd,
            fmt = "none")
```

```python
plt.style.use('default')
y2_sd = np.std(y2)
print(y2_sd)
fig, ax = plt.subplots(figsize =(7,3))
ax.plot(x,y2, color ='red')

ax.errorbar(x, 
            y2,
            yerr= y2_sd,
            fmt = "none",
            ecolor='indigo',
            elinewidth=4,
            capsize =6,
            capthick =4,
            errorevery=2)
# Every 2 datapoints will have a error bar (errorevery =2 )
```
