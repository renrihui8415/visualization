```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

**Make a basic figure with matplotlib**

- 1. Scatter

```python

# 1. Make some random data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# 2. Make a basic scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

- 2. Bar

```python
# 3. Modify the scatter plot to the bar plot
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y)
```

- 3. Line

```python
# 4. Line plot
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

- 4. Hist

```python
# 5. Histogram
fig, ax = plt.subplots(figsize = (5,3))
# Plot the histogram
ax.hist(y, bins=10, edgecolor='black')
```

- 5. Labels and Title

```python
# 6. Add axis labels and titles
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time')
ax.set_xlabel('Years Since Start')
ax.set_ylabel('Total Growth')
```

- 6. Font, Color, Size

```python
font1 = {'family':'sans-serif','color':'blue', 'size':20}
font2 = {'family':'monospace','color':'green', 'size':14}
```

```python
# 6. Add axis labels and titles
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict = font1, loc = 'right')
ax.set_xlabel('Years Since Start', fontdict = font2)
ax.set_ylabel('Total Growth', fontdict = font2, loc = 'bottom')
```

- 7. Modify data points
https://matplotlib.org/stable/api/markers_api.html

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.scatter(x,
           y,
           marker = '*',
           color  = 'indigo')
```

- 8. Change Line Style

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,
        y,
        marker     = '*',
        color      = 'indigo',
        linestyle  = '--',
        linewidth  = 2)
```

- 9. Modify Marker

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,
        y,
        marker     = '*',
        color      = '#7425b9',
        linestyle  = '--',
        linewidth  = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
```

- 10. Grid Line

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,
        y,
        marker     = '*',
        color      = '#7425b9',
        linestyle  = '--',
        linewidth  = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis='y')
```

-  11. Customize the Grid Line

```python
fig, ax = plt.subplots(figsize=(5, 3))

# Plot the data
ax.plot(x,
        y,
        marker='*',
        color='#7425b9',
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')
# Customize the grid lines
ax.grid(axis='y', color='grey', linestyle='-.', linewidth=0.7)
```

Activity with Matplotlib

```python
# Time Series Animation
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.lines import Line2D

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
xdata, ydata = [], []
lines = []  # To store different colored lines

# Initialization function
def init():
    ax.set_xlim(0, 100)
    ax.set_ylim(min(y) - 10, max(y) + 10)
    ax.set_title('Total Growth Over Time')
    ax.set_xlabel('Time')
    ax.set_ylabel('Total Growth')
    return lines

# Update function
def update(frame):
    xdata.append(x[frame])
    ydata.append(y[frame])
    
    # Clear previous lines
    for line in lines:
        line.remove()
    lines.clear()
    
    # Add new segments
    for i in range(1, len(xdata)):
        color = 'red' if ydata[i] > ydata[i - 1] else 'green'
        line_segment = Line2D(
            [xdata[i - 1], xdata[i]],
            [ydata[i - 1], ydata[i]],
            color=color,
            marker='o',
            markersize=5
        )
        ax.add_line(line_segment)
        lines.append(line_segment)
    
    ax.relim()
    ax.autoscale_view()
    return lines

# Create the animation
ani = FuncAnimation(
    fig, update, 
    frames=range(1, len(x)),  # Use the range of x values for frames
    init_func=init, 
    blit=True, 
    interval=100  # Adjust the interval for smooth animation
)


from IPython.display import HTML

HTML(ani.to_html5_video())
```
